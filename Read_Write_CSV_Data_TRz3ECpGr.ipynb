## Step 1: Generate some Random Data

```python
# Define required libraries
import numpy as np
import pandas as pd
```

```python
# Create a dataframe called "random_data" - 4 cols x 100 row 
random_data = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
random_data.head()
```

## Step 2: Use the Project Library to read/write directly to and from the project

##### Documentation: https://www.ibm.com/docs/en/cloud-paks/cp-data/4.0?topic=lib-watson-studio-python#save-data

**There are 2 options for writing data frames to a project**
1. Write the data frame to the file system of the container, and then upload it to the project - shown in this notebooks
2. Write the data frame directly to the project using a buffer object

```python
# Import the lib
from ibm_watson_studio_lib import access_project_or_space
wslib = access_project_or_space()
```

```python
# Option 1: write the dataframe to an intermediate file
random_data.to_csv("WrittenToProjectUpload.csv")

# upload the intermediate file to your Watson Studio project
wslib.upload_file(file_path="WrittenToProjectUpload.csv", asset_name="WrittenToProjectUpload.csv")
```

#### You should see the file in your list of Project Data Assets

## Step 3: Read the data from the project assets and drop the index that is automatically created
